```python
def getTriples(text):
    predictor = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/openie-model.2020.03.26.tar.gz")
    tree = predictor.predict(sentence=text)
    tri_obj = []
    tri_vb = []
    tri_sub = []
    for word in tree['verbs']:
        if '[ARG0' in word['description']:   
            if '[ARG1' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG0:',']')
                obj_index = reduce(obj)
                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[int(obj_index)]
                else:
                    continue

                vb = printX(li,'[V: ',']')
                subj = printX(li, '[ARG1:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    continue
                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)
            if '[ARGM-LOC' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG0:',']')
                obj_index = reduce(obj)


                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[int(obj_index)]
                else:
                    continue

                vb = printX(li,'[V: ',']')
                subj = printX(li, '[ARGM-LOC:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    continue

                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)
            if '[ARGM-TMP' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG0:',']')
                obj_index = reduce(obj)


                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[int(obj_index)]
                else:
                    continue

                vb = printX(li,'[V: ',']')
                subj = printX(li, '[ARGM-TMP:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    continue

                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)

            if '[ARGM-MNR' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG0:',']')
                obj_index = reduce(obj)


                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[int(obj_index)]
                else:
                    continue

                vb = printX(li,'[V: ',']')
                subj = printX(li, '[ARGM-MNR:',']')
                subj_index = reduceRB(subj)

                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj)
            elif '[ARG2' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG0:',']')
                obj_index = reduce(obj)
                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[int(obj_index)]
                else:
                    continue

                vb = printX(li,'[V: ',']')
                subj = printX(li, '[ARG2:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    continue
                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)   
            

        elif '[ARG1' in word['description']:
            if '[ARGM-LOC' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG1: ',']')
                obj_index = reduce(obj)

                if obj_index !=0:
                    if obj_index == (-1):
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[obj_index]
                else:
                    obj1 = 'invalid'
                    continue

                vb = printX(li,'[V: ',']')

                subj = printX(li, '[ARGM-LOC:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    subj1 = 'invalid'
                    continue
                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)
            if 'ARGM-DIR' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG1:',']')
                obj_index = reduce(obj)

                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[obj_index]
                else:
                    obj1 = 'invalid'
                    continue

                vb = printX(li,'[V: ',']')

                subj = printX(li, '[ARGM-DIR:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    continue
                    
                    
                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)
            if 'ARGM-TMP' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG1:',']')
                obj_index = reduce(obj)
                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[obj_index]
                else:
                    obj1 = 'invalid'
                    continue

                vb = printX(li,'[V: ',']')

                subj = printX(li, '[ARGM-TMP:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    continue       
                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)
            if '[ARG2' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG1:',']')
                obj_index = reduce(obj)

                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[obj_index]
                else:
                    obj1 = 'invalid'
                    continue

                vb = printX(li,'[V: ',']')

                subj = printX(li, '[ARG2:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    subj1 = 'invalid'
                    continue
                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)
                
                
            elif '[ARG3' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG1:',']')
                obj_index = reduce(obj)

                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[obj_index]
                else:
                    obj1 = 'invalid'
                    continue

                vb = printX(li,'[V: ',']')

                subj = printX(li, '[ARG3:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    subj1 = 'invalid'
                    continue
                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)
            
                   
            if 'C-ARG1' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG1:',']')
                obj_index = reduce(obj)

                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[obj_index]
                else:
                    obj1 = 'invalid'
                    continue

                vb = printX(li,'[V: ',']')

                subj = printX(li, '[C-ARG1:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    continue
                    
                    
                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)
            
            if 'ARGM-DIR' in word['description']:
                li =  word['description']
                obj = printX(li,'[ARG1:',']')
                obj_index = reduce(obj)

                if obj_index !=0:
                    if obj_index == -1:
                        obj1 = obj
                    else:
                        obj = list(obj.split(" "))
                        obj1 = obj[obj_index]
                else:
                    obj1 = 'invalid'
                    continue

                vb = printX(li,'[V: ',']')

                subj = printX(li, '[ARGM-DIR:',']')
                subj_index = reduce(subj)
                if subj_index != 0:
                    if subj_index == -1:
                        subj1 = subj
                    else:
                        subj = list(subj.split(" "))
                        subj1 = subj[int(subj_index)]
                else:
                    continue
                    
                    
                tri_obj.append(obj1)
                tri_vb.append(vb)
                tri_sub.append(subj1)
                
        
    length = len(tri_vb)
    
    # declaration and adding columns
    triples = []
    for j in range(length):
        column = []
        for i in range(3):
            column.append(0)
        triples.append(column)

    # filling with data
    for i in range(0, length): 
        triples[i][0] = tri_obj[i]
        triples[i][1] = tri_vb[i]
        triples[i][2] = tri_sub[i]
    arr = np.array(triples)
    triples = (np.array(list(set([tuple(t) for t in arr]))))
    return triples
```

```python
def isPerson(st):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(str(st))

    for ent in doc.ents:
        if ent.label_ == "PERSON":
            return 1
    return int(0)
def reduceRB(st):
    pos = isObj(st)
    if 'RB' in pos:
        i = pos.index('RB')
        return i
    else:
        reduce(st)
def printX(s, start,end):
    printx = (s.split(start)[1].split(end)[0])
    return printx

def isObj(st):
    doc = nlp(st)
    pos1 = []
    for token in doc:
        pos1.append(token.tag_)
    return pos1

def reduce(st):
    pos1 = isObj(st)    
    n = pos1.count('NN') + pos1.count('NNS') + pos1.count('NNP') + pos1.count('PRP')
    if n == 1:
        if pos1.count('NN'):
            i = pos1.index('NN')
        elif pos1.count('NNS'):
            i = pos1.index('NNS')
        elif pos1.count('NNP'):
            i = pos1.index('NNP')
        else:
            i = pos1.index('PRP')
        return i
            
    elif n>1:
        # the x of y, return y
        nn = ['NNP','NNP']
        sub1 = ['IN', 'NN', 'NNS']
        sub2 = ['IN', 'NN']
        sub3 = ['IN', 'NNS']
        sub8 = ['IN','NNP']
        sub4 = ['DT', 'JJ', 'NN']
        sub5 = ['DT', 'JJ', 'NNS']
        sub9 = ['DT', 'NN', 'NNP']
        sub19 = ['DT', 'NN', 'NNPS']
        sub26 = ['DT', 'NNP', 'NN']
        sub10 = ['IN','DT', 'NN']
        sub6 = ['DT', 'NN']
        sub7 = ['DT', 'NNS']
        sub11 = ['NNS', 'IN', 'NN']
        sub12 = ['_SP','DT','JJ','NNPS']
        sub13 = ['_SP','DT','JJ','NNP']
        sub14 = [ 'IN','JJ','NN']
        sub15 = [ 'IN','JJ','NNS']
        sub30 = ['IN','JJ','NNP','NN']
        sub16 = ['JJ', 'IN', 'DT', 'NN']
        sub17 = ['JJ', 'IN', 'DT', 'NNS']
        sub18 = ['NNP', 'HYPH', 'NN', 'NN']
        sub = ['NNP', 'HYPH', 'NNP']
        sub0 = ['NNP', 'HYPH', 'NN', 'NNS']
        sub19 = ['JJ', 'NN']
        sub20 = ['JJ', 'NNS']
        sub21 = ['NNP', 'POS', 'NNP', 'NNP']
        sub22 = ['CD', 'NN']
        sub23 = ['CD', 'NNS']
        sub24 = ['NNP', 'POS', 'NN']
        sub25 = ['NNP','NNP','NN']
        sub27 = ['NNP', 'NNP','POS', 'NN']
        sub28 = ['IN','CD']
        sub29 = ['DT', 'NN', 'IN', 'NN']
        sub31 = ['HYPH','IN','HYPH','VBN','NN',]
        sub32 = ['DT', 'NN', 'IN']
        
        if(all(x in pos1 for x in sub18)): 
            i = pos1.index('HYPH')+2
            return i
        elif(all(x in pos1 for x in nn)): 
                i = pos1.index('NNP')
                return i
        elif(all(x in pos1 for x in sub31)): 
                i = pos1.index('HYPH')+4
                return i
        elif(all(x in pos1 for x in sub)): 
            i = pos1.index('NNP')
            return -1
        elif(all(x in pos1 for x in sub27)): 
            i = pos1.index('POS')+1
            return i
        elif(all(x in pos1 for x in sub25)): 
            i = pos1.index('NN')
            return i
        elif(all(x in pos1 for x in sub29)): 
            i = pos1.index('DT')+3
            j = pos1.index('IN')+1
            if i==j:
                return i
            return pos1.index('NN') 
        elif(all(x in pos1 for x in sub32)): 
            i = pos1.index('NN')
            return i
           
        elif(all(x in pos1 for x in sub10)): 
            i = pos1.index('DT')+1
            return i
        elif(all(x in pos1 for x in sub0)): 
            i = pos1.index('NNS')
            return i
        elif(all(x in pos1 for x in sub11)): 
            i = pos1.index('IN') + 1
            return i
        elif 'PRP$' in pos1:
            i = pos1.index('PRP$') + 1
            return i
        elif(all(x in pos1 for x in sub9)): 
            i = pos1.index('NNP')
            return i
        elif(all(x in pos1 for x in sub19)): 
            i = pos1.index('JJ')+1
            return i
        elif(all(x in pos1 for x in sub21)): 
            i = pos1.index('POS')+2
            return i
        elif(all(x in pos1 for x in sub24)): 
            i = pos1.index('POS')+1
            return i
        elif(all(x in pos1 for x in sub26)): 
            i = pos1.index('NN')
            return i
        elif(all(x in pos1 for x in sub6)): 
            i = pos1.index('NN')
            return i
        elif(all(x in pos1 for x in sub7)): 
            i = pos1.index('NNS')
            return i
        elif(all(x in pos1 for x in sub1)):  
            i = pos1.index('NNS')
            return i
        elif(all(x in pos1 for x in sub2)):  
            i = pos1.index('IN') + 1
            return i
        elif(all(x in pos1 for x in sub3)):  
            i = pos1.index('IN') + 1
            return i
        elif(all(x in pos1 for x in sub8)):  
            i = pos1.index('NNP')
            return i 
        elif(all(x in pos1 for x in sub4)): 
            i = pos1.index('JJ') +1
            return i
        elif(all(x in pos1 for x in sub5)): 
            i = pos1.index('JJ') +1
            return i
        elif(all(x in pos1 for x in sub13)): 
            i = pos1.index('NNP')
            return i
        elif(all(x in pos1 for x in sub12)): 
            i = pos1.index('NNS')
            return i
        elif(all(x in pos1 for x in sub14)): 
            i = pos1.index('JJ')+1
            return i
        elif(all(x in pos1 for x in sub15)): 
            i = pos1.index('JJ')+1
            return i
        elif(all(x in pos1 for x in sub30)): 
            i = pos1.index('NNP')+1
            return i
        elif(all(x in pos1 for x in sub16)): 
            i = pos1.index('NN')
            return i
        elif(all(x in pos1 for x in sub17)): 
            i = pos1.index('NNS')
            return i
        elif 'WDT' in pos1:
            i = pos1.index('WDT')-1
            return i
        elif 'WP' in pos1:
            i = pos1.index('WP')-1
            return i
        elif(all(x in pos1 for x in sub19)): 
            i = pos1.index('NN')
            return i
        elif(all(x in pos1 for x in sub20)): 
            i = pos1.index('NNS')
            return i
        elif(all(x in pos1 for x in sub22)): 
            i = pos1.index('NN')
            return i
        elif(all(x in pos1 for x in sub23)): 
            i = pos1.index('NNS')
            return i
        elif(all(x in pos1 for x in sub28)): 
            i = pos1.index('CD')
            return i
        else:
            if 'NN' in pos1:
                i = pos1.index('NN')
                return i
            elif 'NNS' in pos1:
                i = pos1.index('NNS')
                return i
            elif 'NNP' in pos1:
                i = pos1.index('NNP')
                return i
            elif 'CD' in pos1:
                if 'NN' in pos1:
                    return pos1.index('NN')
                elif 'NNS' in pos1:
                    return pos1.index('NNS')
                elif 'NNP' in pos1:
                    return pos1.index('NNP')
                elif 'PRP' in pos1:
                    return pos1.index('PRP')
                else:
                    return -1

    else:
        return 0
```
